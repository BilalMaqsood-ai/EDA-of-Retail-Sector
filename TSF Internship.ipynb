```python
import pandas as pd
import matplotlib.pyplot as plt

ax = plt.gca()
y.plot(kind='bar', stacked=True, ax=ax, width=0.4, position=0 , color=['blue'],fontsize=15,
    figsize= (20,10) )
x.plot(kind='bar', stacked=True, ax=ax, width=0.4, position=0,color=['red'],fontsize=15,
    figsize= (20,10) )
ax.set_xlim(0,10.5)
plt.show()
```

```python
data = data.sort_values(by='Profit', ascending=False)
data.head()
```

```python
data['Profit to Sales ratio'] = data.apply(lambda row: row.Profit / row.Sales, axis=1)
data.head()
```

```python
data = data.sort_values(by='Profit to Sales ratio', ascending=False)
data.head()
```

```python
data.head(100)
```

```python
data5 = data.loc[(data.Profit > 500)]
data5.sort_values(by='Profit to Sales ratio', ascending=False)
data5
```

```python
plt.figure(figsize=(15, 8))
plt.scatter(data5['State'], data5['Profit'], c ="blue")
plt.xlim(-0.5,8.5)
plt.ylim(450,2500)
plt.show()
```
